Import the json Imdb Dataset

```python
df = pd.read_json('movies_dataset.json')
df.head()

```

Answer to the following Analytical questions: 

Question 1: Use Histograms to compare the disribution of movie `imdb_score` column related to movies produced in `France` and `Canada`. 

While creating the histgoram:
- Set the number of bins to 20 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
# Create the histogram for Canada here 
canada_data = df[df['country'] == 'Canada']
ax = canada_data.imdb_score.hist(bins=20, grid=False, edgecolor = 'black')
ax.set_xlabel('Imdb score')
ax.set_ylabel('Count')
ax.set_title('Imdb score for canadian movies')

```

```python
# Create the histogram for France here 
france_data = df[df['country'] == 'France']
ax = france_data.imdb_score.hist(bins=20, grid=False, edgecolor = 'black')
ax.set_xlabel('Imdb score')
ax.set_ylabel('Count')
ax.set_title('Imdb score for french movies')
```

Question 2: Create a histogram to check the distribution of `imdb_score` spent to produce Black&White movies.

While creating the histgoram:
- Set the number of bins to 10 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
#To go back to the tree, click on the Jupyter icon
bw_movies = df[df['color'] == 'Black and White']
ax = bw_movies.imdb_score.hist(bins=10, figsize=(8,6), grid=False, edgecolor = 'black')
ax.set_xlabel('Imdb score')
ax.set_ylabel('Count')
ax.set_title('Imdb score for Black and white movies')
```
